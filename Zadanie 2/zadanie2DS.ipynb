#importy

```python
import pandas as pd
```

#odczyt danych z pliku excelowego

```python
samochody = pd.read_csv("samochody1tys.csv")
samochody
```

#sortowanie po roku produkcji w górę

```python
samochody.sort_values("rok_produkcji", ascending=True)
```

#zmiana nazwy kolumny

```python
samochody.rename(columns = {'rok_produkcji':'production_year'})
```

#usuwanie kolumn

```python
samochody.drop(columns=['model','przebieg'])
```

#wyswietlenie x ostatnich rekordów

```python
samochody.tail(10)
```

#wyciaganie jednej kolumny

```python
samochody['marka']
```

#zliczanie unikalnych pozycji w kolumne

```python
samochody['model'].value_counts()
```

#wyciąganie kolumn z poszczególnymi wartościami

```python
samochody[samochody.marka=="BMW"]
```

#laczenie kolumn(merge) inner - tylko jeśli w obydwu kolumnach są wartości

```python
bmw = samochody[samochody.marka=="BMW"]
wojewodztwo = samochody[samochody.wojewodztwo=="Lubelskie"]
pd.merge(bmw, wojewodztwo, how='inner')
```

#rozlozenie rekordow na kolumny

```python
samochody.pivot(columns='model', values='rodzaj_silnika')
```

#liczenie sredniej z kolumny

```python
samochody.cena.mean()
```
