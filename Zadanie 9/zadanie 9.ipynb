imports

```python
import imutils
import cv2
from IPython.display import Image
```

otwarcie obrazu i zwrócenie jego parametrów

```python
image = cv2.imread("jpg/lampiony.jpg")
(h, w, d) = image.shape
print("width={}, height={}, depth={}".format(w, h, d))
cv2.imshow("Image", image)
cv2.waitKey(0)
Image("jpg/lampiony.jpg")
```

ekstrakcja pikseli

```python
(B, G, R) = image[15, 99]
print("R={}, G={}, B={}".format(R, G, B))
```

ekstrakcja ROI - region of interest


```python
image2 = cv2.imread("jpg/jackie.jpg")
roi = image2[20:280, 260:490]
cv2.imshow("ROI", roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite("jpg/roi.jpg", roi)
Image("jpg/roi.jpg")
```

zmiana rozmiaru obrazka bez proporcji

```python
resized = cv2.resize(image, (200, 200))
cv2.imshow("Fixed Resizing", resized)
cv2.imwrite("jpg/resized.jpg", resized)
Image("jpg/resized.jpg")
```

zmiana rozmiaru (proporcje)

```python
r = 300.0 / w
dim = (300, int(h * r))
resized = cv2.resize(image, dim)
cv2.imshow("Aspect Ratio Resize", resized)
cv2.waitKey(0)
#cv2.imwrite("jpg/resized2.jpg", resized)
Image("jpg/resized2.jpg")
```

zmiana rozmiaru z imultils

```python
resized = imutils.resize(image, width=300)
cv2.imshow("Imutils Resize", resized)
cv2.waitKey(0)
#cv2.imwrite("jpg/imutilsResized.jpg", resized)
Image("jpg/imutilsResized.jpg")
```

obracanie obrazka

```python
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
cv2.imshow("OpenCV Rotation", rotated)
cv2.waitKey(0)
cv2.imwrite("jpg/rotated.jpg", rotated)
Image("jpg/rotated.jpg")
```

obracanie z imultils

```python
rotated = imutils.rotate(image, -45)
cv2.imshow("Imutils Rotation", rotated)
cv2.waitKey(0)
cv2.imwrite("jpg/rotatedImutils.jpg", rotated)
Image("jpg/rotatedImutils.jpg")
```

obracanie z imultis bez obcinania krawędzi

```python
rotated = imutils.rotate_bound(image, 45)
cv2.imshow("Imutils Bound Rotation", rotated)
cv2.waitKey(0)
cv2.imwrite("jpg/imutilsrotated2.jpg", rotated)
Image("jpg/imutilsrotated2.jpg")
```

rozmycie

```python
blurred = cv2.GaussianBlur(image, (11, 11), 0)
cv2.imshow("Blurred", blurred)
cv2.waitKey(0);
cv2.imwrite("jpg/blurred.jpg", blurred)
Image("jpg/blurred.jpg")
```

rysowanie prostokąta

```python
output = image2.copy()
cv2.rectangle(output, (490, 260), (280, 20), (0, 255, 255), 2)
cv2.imshow("Rectangle", output)
cv2.waitKey(0)
cv2.imwrite("jpg/rysowanie.jpg", output)
Image("jpg/rysowanie.jpg")
```

rysowanie koła

```python
output = image2.copy()
cv2.circle(output, (390, 170), 20, (255, 0, 0), -1)
cv2.imshow("Circle", output)
cv2.waitKey(0)
cv2.imwrite("jpg/rysowanieKola.jpg", output)
Image("jpg/rysowanieKola.jpg")
```

rysowanie linii od palców do nosa jackie chana

```python
output = image2.copy()
cv2.line(output, (180, 200), (390, 170), (0, 0, 255), 5)
cv2.imshow("Line", output)
cv2.waitKey(0)
cv2.imwrite("jpg/rysowanieLinii.jpg", output)
Image("jpg/rysowanieLinii.jpg")
```

wstawianie tekstu

```python
output = image2.copy()
cv2.putText(output, "Hey im jackie chan and i kick", (10, 25),
	cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
cv2.imshow("Text", output)
cv2.waitKey(0)
cv2.imwrite("jpg/tekst.jpg", output)
Image("jpg/tekst.jpg")
```
